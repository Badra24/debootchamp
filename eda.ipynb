```python
from faker import Faker
fake = Faker('id_ID')
for _ in range(1000):
    print(fake.phone_number())
```

```python

def generate_user_for_learning():
    """This function generates real-world-like bank customer data."""
    gender = fake.random_element(elements=('M', 'F'))
    if gender == "F":
        first = fake.first_name_female()
        last = fake.last_name_female()
    else:
        first = fake.first_name_male()
        last = fake.last_name_male()
    return {
        "user_id": fake.uuid4(),
        "card_num":fake.bothify(text='##########'),
        "gender": gender,
        "first_name":first,
        "last_name": last,
        "email": fake.random_element(elements=(first, last))+fake.bothify(text='##')+'@'+fake.free_email_domain() if random.random() > 0.1 else None,
        "job": fake.job(),
        "card_type": fake.random_element(elements=('AMEX', 'MasterCard', 'Visa')),
        "address": fake.address(),
        "branch": fake.city(),
        "phone": fake.phone_number(),
        "birthdate": fake.date_of_birth(minimum_age=16, maximum_age=85),
        "country": "ID",
        "balance_usd": round(fake.pyfloat(left_digits=5, right_digits=2, min_value=2999, positive=True), 2)
                  # if random.random() > 0.05 else -1000.00
    }
```

```python
users_df = pd.DataFrame([generate_user_for_learning() for _ in range(5000)])
users_df.head()
```

```python
from datetime import date, timedelta

def daterange(start_date: date, end_date: date):
    days = int((end_date - start_date).days)
    for n in range(days):
        yield start_date + timedelta(n)

start_date = date(2025, 9, 1)
end_date = date(2025, 10, 1)

trxxx = pd.DataFrame()
for single_date in daterange(start_date, end_date):
    trxxx = pd.concat([trxxx, pd.DataFrame([generate_trx(single_date.strftime("%Y-%m-%d")) for _ in range(random.randint(500, 1000))])])
    # print(single_date.strftime("%Y-%m-%d"))
```
